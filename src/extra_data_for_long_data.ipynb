## Визуализация эбдендингов

```python
import pandas as pd
import torch
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_nikita = pd.read_csv('final_bert_sampled_train_df.csv')
embeddings = torch.load('sentence_embeddings.pt')
labels = np.array(df_nikita['labels'].to_list())
```

```python
embeddings.shape
```

```python
pca = PCA(n_components=2)
embeddings_pca = pca.fit_transform(embeddings)
```

```python
# Визуализация эмбеддингов
plt.figure(figsize=(10, 8))
for label in np.unique(labels):
    plt.scatter(embeddings_pca[labels == label, 0], embeddings_pca[labels == label, 1], label=label)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization of Sentence Embeddings')
plt.legend()
plt.show()
```

```python
## TSNE 
```

```python
import numpy as np
from sklearn.manifold import TSNE

X_embedded = TSNE(n_components=2).fit_transform(embeddings)
```

```python
df_embeddings = pd.DataFrame(X_embedded)
df_embeddings = df_embeddings.rename(columns={0:'x',1:'y'})
df_embeddings = df_embeddings.assign(label=df_nikita.labels.values)
```

```python
import plotly.express as px
```

```python
fig = px.scatter(
    df_embeddings, x='x', y='y',
    color='label', labels={'color': 'label'},
    hover_data=['text'], title = 'News Texts Embedding Visualization')
fig.show()
```

## UMAP

```python
import torch
import umap.umap_ as umap
import matplotlib.pyplot as plt
```

```python
umap_embedding = umap.UMAP(n_neighbors=21, min_dist=0.1, metric='cosine').fit_transform(embeddings)
```

```python
# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(umap_embedding[:, 0], umap_embedding[:, 1], c=labels, cmap='viridis', s=10)
plt.colorbar(label='Class')
plt.title('UMAP Visualization of Sentence Embeddings')
plt.show()
```
